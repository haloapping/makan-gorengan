# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Sample Data

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
x.shape
```

# Histogram

## Object Oriented Style

```python
fig, ax = plt.subplots(dpi=125)

ax.hist(x, bins=50, color="g", alpha=0.30)
ax.set_title("Histogram with Object Oriented Style")
ax.grid()
ax.set_xlabel("x")
ax.set_ylabel("Freq")
ax.text(40, 350, "$\mu=100,\ \sigma=15$")

plt.show()
```

## Pyplot Style

```python
plt.figure(dpi=125)
plt.hist(x,
         bins=50,
         facecolor="y",
         alpha=0.50)
plt.title("Histogram with Pyplot Style")
plt.xlabel("x")
plt.ylabel("Freq")
plt.grid()
plt.text(40, 350, "$\mu=100,\ \sigma=15$")
plt.show()
```
