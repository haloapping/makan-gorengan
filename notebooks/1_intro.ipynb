# Matplotlib

- Matplotlib adalah salah satu module yang digunakan untuk visualisasi data dalam bentuk graphs.
- Graphs digambarkan pada suatu `figure` yang di dalamnya terdapat satu atau lebih `axes`.

# Import Modules

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print(f"matplotlib version : {matplotlib.__version__}")
print(f"numpy version      : {np.__version__}")
```

# Simple Plotting

- Siapkan sebuah `figure` dan `axes`. Proses plotting dilakukan pada `axes`.

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 2, 4, 5]

fid, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Dapat juga menggunakan method `plot()` yang terdapat pada pyplot untuk melakukan plotting sederhana.

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 2, 4, 5]

plt.plot(x, y)
plt.show()
```

# Komponen `figure` Matplotlib

<img src="../images/mpl_anatomy.png">

- `figure` adalah container atau canvas yang merupakan tempat berada sekumpulan `axes`.
- `axes` adalah media atau tempat untuk menampilkan hasil plotting. Jumlah `axes` dapat lebih dari satu untuk sebuah `figure`.
- `axis` adalah sumbu pada hasil plotting di setiap `axes`, yakni sumbu x, y, dan z.

## `figure`

```python
fig = plt.figure()
plt.show()
```

- Menghasilkan sebuah object `figure` berukuran 432x288 tanpa `axes`.
- Tidak menampilkan hasil plotting, karena variabel `fig` hanya mengembalikan object.

## `axes`

```python
fig, ax = plt.subplots()
plt.show()
```

Membuat sebuah `figure` dengan 1 `axes`

```python
fig, ax = plt.subplots(2, 3)
plt.show()
```

# Cara menggunakan matplotlib

## OO Style (Object Oriented Style)

Secara eksplisit membuat `figures` beserta `axes` dan memanggil methods dari keduanya.

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots()

ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")

ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
ax.legend()
ax.grid()
plt.show()
```

## Pyplot Style

- Tidak secara eksplisit membuat figures beserta axes dan memanggil methods dari keduanya.
- Plotting dapat dlakukan langsung dengan menggunakan fungsi pada `pyplot`.

```python
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label="qubic")
plt.title("Simple Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```
