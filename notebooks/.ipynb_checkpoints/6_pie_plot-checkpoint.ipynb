# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt

print(f"Matplotlib version: {matplotlib.__version__}")
```

# Simple Pie Plot

```python
data = [23, 43, 34, 50]
kategori = ["P", "I", "N", "G"]

plt.figure(dpi=140)
plt.pie(data,
        labels=kategori,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Simple Pie Plot")
plt.show()
```

# Color, Legend, and Explode

## Color

```python
data = [23, 43, 34, 50]
kategori = ["P", "I", "N", "G"]
warna = ["yellow", "green", "red", "pink"]

plt.figure(dpi=140)
plt.pie(data,
        colors=warna,
        labels=kategori,
        autopct="%1.1f%%")
plt.title("Plot with Color")
plt.show()
```

## Legend

```python
plt.figure(dpi=140)
plt.pie(data,
        colors=warna,
        labels=kategori,
        autopct="%1.1f%%")
plt.title("Plot with Color")
plt.legend()
plt.show()
```

## Explode

```python
explode_var = [0, 0, 0.2, 0.3]

plt.figure(dpi=140)
plt.pie(data,
        colors=warna,
        labels=kategori,
        autopct="%1.1f%%",
        explode=explode_var,
        startangle=90)
plt.title("Plot with Color")
plt.legend()
plt.show()
```
