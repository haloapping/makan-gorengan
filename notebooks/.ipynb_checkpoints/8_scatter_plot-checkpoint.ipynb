# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt

print(f"Matplotlib version: {matplotlib.__version__}")
```

# Simple Scatter Plot

```python
# Data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [4, 5, 6, 7, 8, 9, 15, 20]

# Visualize
plt.figure(dpi=100)
plt.scatter(x, y, label="data-1", color="gold")
plt.title("Simple Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

# Multiple Scatter Plot

```python
# Data
x_1 = [1, 2, 3, 4, 5, 6, 7, 8]
y_1 = [4, 5, 6, 7, 8, 9, 15, 20]

x_2 = [4, 5, 6, 7, 8, 9, 15, 20]
y_2 = [1, 2, 3, 4, 5, 6, 7, 8]

# Visualize
plt.figure(dpi=100)
plt.scatter(x_1, y_1, label="data-1", color="green")
plt.scatter(x_2, y_2, label="data-2", color="gold")
plt.title("Multiple Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

# Marker Config

```python
# Data
x_1 = [1, 2, 3, 4, 5, 6, 7, 8]
y_1 = [4, 5, 6, 7, 8, 9, 15, 20]

x_2 = [4, 5, 6, 7, 8, 9, 15, 20]
y_2 = [1, 2, 3, 4, 5, 6, 7, 8]

# Visualize
plt.figure(dpi=100)
plt.scatter(x_1,
            y_1,
            label="data-1",
            color="darkkhaki",
            marker="^",
            linewidths=2,
            edgecolor="orangered",
            s=200)

plt.scatter(x_2,
            y_2,
            label="data-2",
            color="lime",
            marker="s",
            linewidths=2,
            edgecolor="seagreen",
            s=100)

plt.title("Multiple Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

# Scatter Plot untuk Multi Dimensi

```python
# Import Dataset
from sklearn.datasets import load_iris
iris = load_iris()

# Feature and Target
features = iris["data"].T
target = iris["target"]
feature_names = iris["feature_names"]

# Visualize
plt.figure(dpi=100)

plt.scatter(features[0],
            features[1],
            alpha=0.2,
            s=features[3] * 100,
            c=target,
            cmap="viridis")

plt.title("Iris Dataset")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.colorbar()
plt.grid()
plt.show()
```
