# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Sample Dataset

```python
n = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)
```

# Contour Plot without Color Fill

```python
plt.figure(dpi=100)
plt.contour(x, y, z)
plt.colorbar()
plt.title("Contour Plot without Color Fill")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Contour Plot with Color Fill

```python
plt.figure(dpi=100)
plt.contourf(x, y, z)
plt.colorbar()
plt.title("Contour Plot with Color Fill")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
