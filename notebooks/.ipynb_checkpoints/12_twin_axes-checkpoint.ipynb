# Import Modules

```python
# %matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Twin Axes 1

```python
# Dataset
x = np.linspace(1, 10, 25)

# Visualize
fig, ax_1 = plt.subplots()
fig.dpi = 100
ax_1.plot(x, np.exp(x), "bs-", label="exp")
ax_1.set(xlabel="x-axis",
         ylabel="exp")

ax_2 = ax_1.twinx()
ax_2.plot(x, np.log(x), "ro-", label="log")
ax_2.set(ylabel="log")

fig.suptitle("Twin Axes")
fig.legend(loc="upper right")
plt.show()
```

# Twin Axes 2

```python
x = np.random.randint(low=18, high=50, size=40)

def celcius_to_farenheit(celcius=0):
    return celcius * 1.8 + 32

def axis_convertion(ax_1):
    y1, y2 = ax_1.get_ylim()
    ax_2.set_ylim(celcius_to_farenheit(y1), celcius_to_farenheit(y2))
    ax_2.figure.canvas.draw()

fig, ax_1 = plt.subplots()
fig.dpi = 100
ax_2 = ax_1.twinx()

ax_1.callbacks.connect("ylim_changed", axis_convertion)
ax_1.plot(x)
ax_1.set(xlabel="Hari",
         ylabel="Celcius")

ax_2.set(ylabel="Farenheit")

fig.suptitle("Temperatur Udara")

plt.show()
```
