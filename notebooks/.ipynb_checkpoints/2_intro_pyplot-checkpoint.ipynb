# Pyplot

- Salah satu module yang digunakan untuk melakukan plotting dengan menggunakan style MATLAB.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Simple Plotting

```python
plt.plot([2, 5, 7, 11])
plt.ylabel("y-axis")
plt.show()
```

Jika pemanggilan `plot()` hanya melewatkan satu parameter, maka nilai tersebut berasosiasi dengan sumbu y dan sumbu x akan secara otomatis menggunakan deret angka dimulai dari 0.

```python
plt.plot([2, 5, 7, 11], [1, 2, 10, 4])
plt.show()
```

# Pengaturan Format pada Plot

- Selain melewatkan dua nilai ke parameter, yaitu data pada sumbu x dan y, kita juga dapat melewatkan parameter warna, bentuk plot, dan nilai minimal serta maksimal pada setiap `axis`.

## Warna Plot

```python
plt.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6], "g")
plt.show()
```

## Bentuk Plot

```python
plt.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6], "g--")
plt.show()
```

## Nilai minimal dan maksimal untuk setiap `axis`

```python
plt.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6], "go")
plt.axis([0, 10, 0, 10])
plt.show()
```

# Multiplot

```python
t = np.arange(0, 5, 0.2)
t
```

## Multiple Plot

```python
plt.plot(t, t, "r--")
plt.plot(t, t**2, "bs")
plt.plot(t, t**3, "g^")
plt.show()
```

## Single Plot

```python
plt.plot(t, t, "r--",
         t, t**2, "bs",
         t, t**3, "g^")
plt.show()
```

# Plotting dengan Keywords

- Pada Matplotlib dimungkinkan untuk melakukan plotting dengan menggunakan dengan struktur data yang disertai dengan argumen, seperti `dictionary`.

```python
data ={"a": np.arange(50),
       "c": np.random.randint(0, 50, 50),
       "d": np.random.randn(50)}

data["b"] = data["a"] + 10 * np.random.randn(50)
data["d"] = np.abs(data["d"]) * 100
```

## Scatter Plot

```python
plt.scatter("a", "b", data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

## Scatter Plot with color and size parameter

```python
plt.scatter("a", "b", c="c", s="d", data=data, marker="o")
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

# Plotting untuk data kategorikal

```python
name = ["Andi", "Lehar", "Apping"]
weight = [60, 50, 55]
```

## Line Plot

```python
plt.plot(name, weight)
plt.title("Sebaran Berat Badan")
plt.xlabel("Nama")
plt.ylabel("Berat Badan (kg)")
plt.show()
```

## Scatter Plot

```python
plt.scatter(name, weight)
plt.title("Sebaran Berat Badan")
plt.xlabel("Nama")
plt.ylabel("Berat Badan (kg)")
plt.show()
```

## Bar Plot

```python
plt.bar(name, weight)
plt.title("Sebaran Berat Badan")
plt.xlabel("Nama")
plt.ylabel("Berat Badan (kg)")
plt.show()
```
