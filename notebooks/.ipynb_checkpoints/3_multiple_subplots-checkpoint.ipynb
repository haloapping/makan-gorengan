# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Simple Line Plot

```python
x = np.arange(0, 2, 0.01)
s = np.sin(2 * np.pi * x)
```

## Object Oriented Style (OO Style)

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, s)
ax.set(title="Visualisasi nilai sin",
       xlabel="nilai x",
       ylabel="nilai sin(x)")
ax.grid()
plt.show()
```

# Multiple Subplots

```python
x1 = np.linspace(0, 5, 100)
x2 = np.linspace(0, 2, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

## Object Oriented Style  (OO Style)

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))
fig.suptitle("Multiple Subplots")

ax1.plot(x1, y1, "y")
ax1.set(ylabel="Nilai $cos(x_1)$")

ax2.plot(x2, y2, "g")
ax2.set(ylabel="Nilai $cos(x_2)$")
plt.show()
```

## Pyplot

```python
from matplotlib.pyplot import figure
figure(figsize=(10, 6))

plt.subplot(2, 1, 1, )
plt.title("Multiple Subplots")

plt.plot(x1, y1, "y")
plt.ylabel("Nilai $cos(x_1)$")

plt.subplot(2, 1, 2)
plt.plot(x2, y2, "g")
plt.ylabel("Nilai $cos(x_2)$")

plt.show()
```
