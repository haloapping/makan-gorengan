# Import Modules

```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# 3D Axes

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.set(title="3D Axes",
       xlabel="x-axis",
       ylabel="y-axis",
       zlabel="z-axis")
plt.show()
```

# 3D Line Plot

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

fig = plt.figure()
ax = plt.axes(projection="3d")

ax.plot(x, y, z, color="green")
ax.set(title="3D Line Plot",
       xlabel="x-axis",
       ylabel="y-axis",
       zlabel="z-axis")

plt.show()
```

# 3D Scatter Plot

```python
z = np.linspace(-5, 0, 1000)
x = z * np.sin(100 * z)
y = z * np.cos(100 * z)

fig = plt.figure()
ax = plt.axes(projection="3d")

ax.scatter(x, y, z, color="gold")
ax.set(title="3D Line Plot",
       xlabel="x-axis",
       ylabel="y-axis",
       zlabel="z-axis")

plt.show()
```
