# Barplot

Digunakan untuk memvisualisasikan frekuensi kemunculan data bertipe numerik atau diskrit.

# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Simple Bar Plot

```python
kategori = ["P", "I", "N", "G"]
freq = [80, 100, 150, 55]

plt.figure(dpi=140)
plt.bar(kategori, freq)
plt.title("Simple Bar Plot")
plt.xlabel("Kategori")
plt.ylabel("Freq")
plt.grid()
plt.show()
```

# Pengaturan Color and Grid

## Color

```python
kategori = ["P", "I", "N", "G"]
freq = [80, 100, 150, 55]

plt.figure(dpi=140)
plt.bar(kategori, freq, color="yellow", alpha=0.5)
plt.title("Color Configuration")
plt.ylabel("Freq")
plt.xlabel("Kategori")
plt.show()
```

## Grid

```python
kategori = ["P", "I", "N", "G"]
freq = [80, 100, 150, 55]

plt.figure(dpi=140)
plt.bar(kategori, freq, color="green", alpha=0.5)
plt.title("Grid Configuration")
plt.ylabel("Freq")
plt.xlabel("Kategori")

plt.grid(axis="y",
         linestyle="--",
         linewidth=2,
         alpha=0.5)

plt.show()
```

# Horizontal Bar Plot

```python
kategori = ["P", "I", "N", "G"]
freq = [80, 100, 150, 55]

plt.figure(dpi=140)
plt.barh(kategori, freq)
# plt.barh(kategori[::-1], freq[::-1])
plt.title("Horizontal Bar Plot")
plt.ylabel("Kategori")
plt.xlabel("Freq")

plt.grid(axis="x",
         linestyle="--",
         linewidth=2,
         alpha=0.5)

plt.show()
```

# Stacked Bar Plot

```python
freq_1 = [10, 20, 30, 40]
freq_2 = [30, 40, 50, 60]
kategori = ["P", "I", "N", "G"]

plt.figure(dpi=140)
plt.bar(kategori, freq_1, label="freq 1")
plt.bar(kategori, freq_2, label="freq 2", bottom=freq_1)

plt.grid(axis="y",
         linestyle="--",
         alpha=0.5)
plt.legend()
plt.title("Stacked Bar")
plt.xlabel("Kategori")
plt.ylabel("Freq")
plt.show()
```

# Grouped Bar Plot

```python
freq_1 = [10, 20, 30, 40]
freq_2 = [30, 40, 50, 60]
kategori = ["P", "I", "N", "G"]
x = np.arange(len(kategori))
width = 0.35

plt.figure(dpi=140)
plt.bar(x-width/2, freq_1, width, label="freq 1")
plt.bar(x+width/2, freq_2, width, label="freq 2")
plt.title("Grouped Bar")
plt.xlabel("Kategori")
plt.ylabel("Freq")
plt.grid(axis="y")
plt.xticks(x, kategori)
plt.show()
```
