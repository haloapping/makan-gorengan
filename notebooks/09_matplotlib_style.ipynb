# Import Module

```python
from matplotlib import style

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# List of Style

```python
def visualize_barplot(style_name: str = None):
    data_1 = [25, 80, 40, 40]
    data_2 = [36, 54, 88, 90]
    kategori = list("PING")
    x = np.arange(len(kategori))
    width = 0.35
    
    plt.bar(x-width/2, data_1, width, label="data-1")
    plt.bar(x+width/2, data_2, width, label="data-2")
    plt.title(f"Grouped Bar ({style_name})")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.xticks(x, kategori)
    plt.grid(linestyle="--", linewidth=1, axis="y", alpha=0.75)
    plt.legend()
    plt.show()
```

```python
visualize_barplot()
```

```python
plt.style.available
```

```python
for style_name in plt.style.available:
    style.use(style_name)
    plt.figure(dpi=100)
    visualize_barplot(style_name)
```
