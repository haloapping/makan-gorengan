# Animation

```python
fig = plt.figure()
plt.xlim(0, 4)
plt.ylim(-2, 2)
plt.title("Animation Plot")

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    
    return line

anim = FuncAnimation(fig=fig, func=animate, frames=200, interval=20)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```

```python
f = r"../images/animation.gif" 
writergif = animation.PillowWriter(fps=30) 
anim.save(f, writer=writergif)
```
