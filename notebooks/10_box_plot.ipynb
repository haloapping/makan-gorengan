# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
```

# Simple Box Plot

```python
plt.figure(dpi=120)
plt.boxplot(x=data, labels=["data"])
plt.title("Simple Box Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

# Notched Box Plot

```python
plt.figure(dpi=120)
plt.boxplot(x=data, labels=["data"], notch=True)
plt.title("Simple Box Plot with Notched")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

# Marker Config for Outlier

## Hide Outliers

```python
plt.figure(dpi=120)
plt.boxplot(x=data,
            labels=["data"],
            notch=True,
            showfliers=False)
plt.title("Simple Box Plot with Notched")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

## Marker

```python
plt.figure(dpi=120)
plt.boxplot(x=data,
            labels=["data"],
            notch=True,
            flierprops={"markerfacecolor": "r", "marker": "^"})
plt.title("Simple Box Plot with Notched")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

# Horizontal Box Plot

```python
plt.figure(dpi=120)
plt.boxplot(x=data,
            labels=["data"],
            notch=True,
            vert=False,
            flierprops={"markerfacecolor": "r", "marker": "^"})
plt.title("Simple Box Plot with Notched")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

# Multiple Box Plot

```python
np.random.seed(42)

data_1 = np.random.normal(loc=100, scale=10, size=100)
data_2 = np.random.normal(loc=100, scale=20, size=100)
data_3 = np.random.normal(loc=100, scale=25, size=100)
data_4 = np.random.normal(loc=100, scale=50, size=100)

all_data = [data_1, data_2, data_3, data_4]
labels = ["data-1", "data-2", "data-3", "data-4"]
```

```python
plt.figure(dpi=120)
plt.boxplot(x=all_data,
            labels=labels,
            notch=True,
            vert=False,
            flierprops={"markerfacecolor": "r", "marker": "^"})
plt.title("Simple Box Plot with Notched")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(axis="x")
plt.show()
```
