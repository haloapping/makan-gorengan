# Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version : {matplotlib.__version__}")
print(f"Numpy version      : {np.__version__}")
```

# Sample Dataset

```python
np.random.seed(42)
data = np.random.normal(loc=100, scale=10, size=200)
```

# Simple Violin Plot

```python
plt.figure(dpi=100)
plt.violinplot(data)
plt.title("Simple Violin Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Violin Plot Configuration

```python
plt.figure(dpi=100)
plt.violinplot(dataset=data,
               showextrema=True, # line min and max value
               showmeans=True, # line mean
               showmedians=True, # line median
               quantiles=[0.25, 0.5, 0.75]) # Q1, Q2, Q3
plt.title("Violin Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Horizontal Violin Plot

```python
plt.figure(dpi=100)
plt.violinplot(dataset=data,
               vert=False,
               quantiles=[0.05, 0.95])
plt.title("Horizontal Violin Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Multiple Violin Plot

```python
np.random.seed(42)
data_1 = np.random.normal(loc=100, scale=10, size=200)
data_2 = np.random.normal(loc=130, scale=30, size=200)
data_3 = np.random.normal(loc=80, scale=20, size=200)
data_4 = np.random.normal(loc=60, scale=25, size=200)
all_data = [data_1, data_2, data_3, data_4]
labels = ["data-1", "data-2", "data-3", "data-4"]

plt.figure(dpi=100)
plt.violinplot(dataset=all_data)
plt.title("Multiple Violin Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks([float(i) for i in range(1, 5)], labels=labels)
plt.show()
```
