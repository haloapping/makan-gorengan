# Import Modules

```python
%matplotlib notebook

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f"Matplotlib version: {matplotlib.__version__}")
print(f"Numpy version: {np.__version__}")
```

# Sample Dataset

```python
n = np.linspace(-3, 3, 100)
x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)
```

# 3D Wireframe

```python
fig = plt.figure()

ax = plt.axes(projection="3d")
ax.plot_wireframe(x, y, z, color="orange")
ax.set(title="3D Wireframe",
       xlabel="x-axis",
       ylabel="y-axis",
       zlabel="z-axis")

plt.show()
```

# 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

# ax.plot_surface(x, y, z, color="orange")
ax.plot_surface(x, y, z, cmap="viridis")
ax.set(title="3D Surface",
       xlabel="x-axis",
       ylabel="y-axis",
       zlabel="z-axis")

plt.show()
```
